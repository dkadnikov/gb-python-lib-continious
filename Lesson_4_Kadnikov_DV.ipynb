### Урок 4. Оценка и интерпретация полученной модели. Обсуждение курсового проекта.

#### 1. Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

В деревьях принятия решений, чтобы не допустить переобучения, мы можем ограничить или ругеляыризировать или сделать останов:

— Максимальноую глубину дерева (max_depth)

— Минимальное число объектов в листе (min_samples_split / min_samples_leaf — параметр min_samples_split будет оценивать количество выборок в узле, и если это число меньше минимального, то разделения можно будет избежать, и узел будет листовым. Параметр min_samples_leaf проверяет перед созданием узла, то есть если возможное разделение приводит к появлению дочернего элемента с меньшим количеством выборок, то разделение будет предотвращено (так как минимальное количество выборок для дочернего элемента, являющегося листом, не достигнуто) и узел будет заменен листом.

— Максимальное количества листьев в дереве (max_features)

— Сделать останов в случае, если все объекты в листе относятся к одному классу. 
— Задать требование, что функционал качества при дроблении улучшался как минимум на s процентов.

Соответственно штрафуем мы за излишнюю глубину дерева, слишком большой лист и т.д.
С помощью грамотного выбора подобных критериев и их параметров можно существенно повлиять на качество дерева. 
Тем не менее, такой подбор является трудозатратным и требует проведения кросс-валидации. А еще можно воспользоваться стрижкой. 

 

#### 2. По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

feature_importance (Важность признака - это то, какое влияние данных признак оказывает при прогнозировании целевой переменной. рассчитывается с учтетом его значения у каждого дерева. Т.е. feature_importance - это отношение между количеством выборок, направляемых в узел принятия решения с участием этого признака в любом из деревьев ансамбля, и общим количеством выборок в обучающем наборе.
