### Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных.

#### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.
Precision используется, когда важно не ошибиться при обнаружении искомых объектов.
- При оценке экспресс теста диагноза COVID лучше максимизировать Precision, так как цена положительных ответов наиболее важна при постановке диагноза при экспресс тесте.
- При оценке внутри банковской системы подозрительных операций по карте лучше максимизировать Precision, так как важно найти подозрительные операции и не выбрать операции «нормальные» для пользователя карты.
Recall использутеся, когда нам важно найти максимальное количество объектов искомого класса. 
- При оценки волатильности портфеля акций лучше максимизировать Recall, для определения политики диверсификации с целью снижения рисков портфеля. 
- При оценке спорных нарушений ПДД  с камеры лучше максимизировать Recall, так как можно позволить ошибки определения, но не пропустить нарушение.

#### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?
- F-мера позволяет комплексно оценивать как Precision(точность)  и Recall (полнота), а так же регулировать величину значения вклада в этот показатель Precision  и Recall при потребности.
