### Урок 2. Анализ данных и проверка статистических гипотез.

##### В чём различие между зависимыми и независимыми выборками?

- Независимые выборки - это когда из генеральной выборки сравниваются две разные группы данных, сформированные случайным образом, либо относительно некоего признака, например, пола — (мужчины и женщины или молодые и пожилые и т.д.)

- Зависимые выборки как правило возникают, когда из генеральной выборки мы сравниваем две группы в которых данные связаны каким-либо признаком, например когда один человек в первой группе по какому-либо содержательному признаку сопоставляется с соответствующим человеком во второй группе, например, муж из первой группы сравнивается со своей женой, которая во второй группе

##### Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

- Главное условие применения параметрических критериев – наличие конкретного типа распределения генерльной выборки или совокупности, нормальность распределения анализируемых переменных.(Например, для определения параметрического F-критерия равенства дисперсий Фишера, как следует из названия, требуется знание дисперсий, что отсылает нас к знанию закона распределения.)

- Непараметрические аналоги - базируется на предположении о типе распределения генеральной совокупности и не использует параметры этой совокупности. Поэтому для непараметрических критериев используют такой термин, как «критерий, свободный от распределения». (Например, критерий Хи-квадрат Пирсона оперирует частотами наблюдаемы параметров, что не требует знания закона распределения.)
