### Урок 3. Построение модели классификации.

#### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

На примере F1-score:

- micro возвращает функцию которая вычисляет f1 учитывая общее количество true positives, false negatives and false positives результатов (независимо от прогноза для каждого класса в наборе данных) т.е. все выборки в равной степени вносят вклад в итоговую усредненную метрику.
- macro возвращает функцию которая вычисляет f1 для каждого класса и возвращает среднее значение без учета доли для каждого класса в наборе данных. Т.е. все классы в равной степени вносят вклад в итоговую усредненную метрику.
- weighted возвращает функцию для вычисления f1 для каждого класса и возвращает среднее значение с учетом доли каждого класса в наборе данных. Выходит, что чем больше элементов имеет класс, тем важнее f1 для этого класса в вычислении или вклад каждого класса в среднее значение взвешивается по его размеру

Микро - и макро- будут вычислять несколько разные вещи, и, следовательно, их интерпретация отличается. 
Макро-среднее вычислит метрику независимо для каждого класса, а затем возьмет среднее (следовательно, рассматривая все классы одинаково), в то время как микро-среднее будет суммировать вклады всех классов для вычисления средней метрики. В многоклассовой классификации предпочтительнее использовать микро-среднее значение, если вы подозреваете, что может быть дисбаланс классов (т.е. у вас может быть гораздо больше примеров одного класса, чем других классов).
##### Ну и какой тип усреднения выбрать? Как обычно, это во многом зависит от проблемы, которую Мы хотим решить. 

Например у нас есть несбалансированный по классам набор данных? Является ли один класс более важным для правильного усвоения, чем другие? Если у нас есть недопредставленный класс, который важен для вашей проблемы, макросреднение может быть лучше, так как оно будет в равной степени подчеркивать производительность модели для всех классов. 
С другой стороны, если предположение о том, что все классы одинаково важны, неверно, макросреднение будет чрезмерно подчеркивать низкую производительность в нечастом классе. 
Микро-усреднение может быть предпочтительным в настройках с несколькими метками, включая многоклассовую классификацию, где класс большинства должен игнорироваться.
Существует также стратегия усреднения smaples - “выборок”, которая применяется только к задачам с несколькими метками (applies only to multilabel problems. It does not calculate a per-class measure, instead calculating the metric over the true and predicted classes for each sample in the evaluation data, and returning their (sample_weight-weighted) average)


#### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

Математически:
LightGBM предлагает одностороннюю выборку на основе градиента (GOSS)
Catboost предлагает новый метод, называемый выборкой с минимальной дисперсией (MVS)
XGBoost не использует никаких методов взвешенной выборки, что делает его процесс разделения медленнее по сравнению с GOSS и MVS.

А так позволю себе сравнить немного перелопатив много литературы: 
CatBoost работает хорошо только тогда, когда в данных есть категориальные переменные, и мы должным образом настроили их.
XGBoost - его глобальная проблема заключается в том, что он работает слишком медленно, включая время настройки/подьбора его параметров.
Light GBM - плохо с точки зрения точности, зато он самый быстрый. 

Если выбирать среди моделей - то в задачах, где нам нужна максимальная точность, мы можем обучить все алгоритмы, чтобы найти тот, у которого самый высокий AUC для этого конкретного набора данных. Но все пакеты постоянно обновляются и улучшаются.
